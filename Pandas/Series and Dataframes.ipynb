# Series:
------------------------------------
#### Series is one dimensional(1-D) array defined in pandas that can be used to store any data type.

## Here, Data can be:
-------------------------
- A Scalar value which can be integerValue, string
- A Python Dictionary which can be Key, Value pair
- A Ndarray

```python
# When Data contains scalar values

import pandas as pd  

Data =[1, 3, 4, 5, 6, 2, 9]  
a = pd.Series(Data)  
a
```

```python
import pandas as pd  

# Numeric data
Data =[1, 3, 4, 5, 6, 2, 9]  

# predefined index values
Index =['a', 'b', 'c', 'd', 'e', 'f', 'g'] 
  
# Creating series with predefined index values
si = pd.Series(Data, Index) 
si
```

```python
# When Data contains Dictionary

dictionary ={'a':1, 'b':2, 'c':3, 'd':4, 'e':5} 

# Creating series of Dictionary type
sd = pd.Series(dictionary) 
sd
```

```python
sd.to_csv(r'data/series.csv')
```

```python
# When Data contains Ndarray

# Defining 2darray
Data =[[2, 3, 4], [5, 6, 7]]  
  
# Creating series of 2darray
snd = pd.Series(Data)   
snd
```

# DataFrames:
----------------------------------
#### DataFrames is two-dimensional(2-D) data structure defined in pandas which consists of rows and columns.

## Here, Data can be:
--------------------------
- One or more dictionaries
- One or more Series
- 2D-numpy Ndarray

```python
# When Data is Dictionaries

# Define Dictionary 1
dict1 ={'a':1, 'b':2, 'c':3, 'd':4}   
  
# Define Dictionary 2     
dict2 ={'a':5, 'b':6, 'c':7, 'd':8, 'e':9} 
  
# Define Data with dict1 and dict2
Data = {'first':dict1, 'second':dict2} 
  
# Create DataFrame 
df = pd.DataFrame(Data)  

df
```

```python
# When Data is Series

import pandas as pd
  
# Define series 1
s1 = pd.Series([1, 3, 4, 5, 6, 2, 9])   
  
# Define series 2       
s2 = pd.Series([1.1, 3.5, 4.7, 5.8, 2.9, 9.3]) 
  
# Define series 3
s3 = pd.Series(['a', 'b', 'c', 'd', 'e'])     
  
# Define Data
Data ={'first':s1, 'second':s2, 'third':s3} 
  
# Create DataFrame
dfseries = pd.DataFrame(Data)              
dfseries
```

```python
dfseries.to_csv(r'data/dataframe.csv', index = False, header = True)
```

```python
dfseries.isnull()
```

```python
# Program to create DataFrame from 2D array
  
# Import Library
import pandas as pd 
  
# Define 2d array 1
d1 =[[2, 3, 4], [5, 6, 7]] 
  
# Define 2d array 2
d2 =[[2, 4, 8], [1, 3, 9]] 
  
# Define Data
Data ={'first': d1, 'second': d2}  
  
# Create DataFrame
df2d = pd.DataFrame(Data)    
df2d
```

```python
df2d.to_csv(r'data/df2d.csv', index = False, header = True)
```
